```python
import mido
import os
import numpy as np


#import rtmidi
```

```python
path = '/home/ritik/Desktop/music_generation_model/Pop_Music_Midi/'
```

```python
max = -1000
for file in os.listdir(path):
    count = 0
    msg = mido.MidiFile(path + file)
    for i in msg:
        if i.type == 'note_on':
            count = count + 1
            print(i.bytes)
    if max < count:
        max = count
print(max)
```

```python
datax = []
time = 0.0
previous = 0.0
for file in os.listdir(path):
    notes = mido.MidiFile(path + file)
    
    for msg in notes: 
        time += msg.time
        if(msg.type == 'note_on'):
            bit = msg.bytes()
            bit = bit[1:]
            bit.append(msg.channel)
            bit.append(time - previous)
            previous = time
            datax.append(bit)
print(len(datax))
print(datax)    
```

```python
datax = np.array(datax)
mean = np.mean(datax,axis = 0)
std = np.std(datax,axis = 0)
print(mean)
print(std)
datax = (datax[:] - mean) / std
print(datax, len(datax))
```

```python
n_prev = 3
trainx = []
trainy = []
for i in range(0, len(datax) - n_prev):
    
    inter = datax[i:i+n_prev]
    inter2 = datax[i+n_prev]
    trainx.append(inter)
    trainy.append(inter2)

trainx = np.array(trainx)
trainy = np.array(trainy)

print(trainx.shape)
```

```python
np.save('/home/ritik/Desktop/music_generation_model/x_train', trainx)
np.save('/home/ritik/Desktop/music_generation_model/y_train', trainy)
```
